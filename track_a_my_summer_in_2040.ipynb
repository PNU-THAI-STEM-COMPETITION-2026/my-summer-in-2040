## 🌍 Track A: Climate Data Loading
This track uses historical daily meteorological data (1981–2026) for selected locations in Thailand and South Korea.

### Data Source Details:
* **Repository:** [my-summer-in-2040](https://github.com/PNU-THAI-STEM-COMPETITION-2026/my-summer-in-2040)
* **Data Format:** Multiple daily CSV weather files located in `data/THA/` and `data/KOR/` folders.
* **Important Note:** The CSV files contain column headers directly on the first line. There are no metadata header lines to skip, and there are no missing values (all incomplete or -999.0 values have been resolved/removed).

```python
# In Google Colab, we need to clone the repository to get the data files.
# If running locally in the cloned repository, this clone step will be skipped.
if not os.path.exists("my-summer-in-2040") and not os.path.exists("data"):
    print("Cloning Track A repository...")
    !git clone https://github.com/PNU-THAI-STEM-COMPETITION-2026/my-summer-in-2040.git
else:
    print("Data repository folder found!")

# Determine the base directory
base_dir = "my-summer-in-2040" if os.path.exists("my-summer-in-2040") else "."

# ----------------------------------------------------
# Load ALL Climate CSV Data files (Thailand & Korea)
# ----------------------------------------------------
tha_files = glob.glob(f"{base_dir}/data/THA/*.csv")
kor_files = glob.glob(f"{base_dir}/data/KOR/*.csv")

all_climate_dfs = []

# Load Thailand data
for file_path in tha_files:
    loc_name = os.path.basename(file_path).replace("Single_point-", "").replace(".csv", "")
    df = pd.read_csv(file_path)
    df["Location"] = loc_name
    df["Country"] = "Thailand"
    all_climate_dfs.append(df)

# Load Korea data
for file_path in kor_files:
    loc_name = os.path.basename(file_path).replace("Single_point-", "").replace(".csv", "")
    df = pd.read_csv(file_path)
    df["Location"] = loc_name
    df["Country"] = "South Korea"
    all_climate_dfs.append(df)

if all_climate_dfs:
    # Combine all locations into a single master DataFrame
    df_climate_all = pd.concat(all_climate_dfs, ignore_index=True)
    print("\n--- All Climate Data Loaded & Merged ---")
    print(f"Total files loaded: {len(all_climate_dfs)}")
    print(f"Combined Shape: {df_climate_all.shape}")
    print(f"Available Locations: {df_climate_all['Location'].unique()}")
    display(df_climate_all.head())
else:
    print("\nNo climate files found. Please verify the paths.")
```
